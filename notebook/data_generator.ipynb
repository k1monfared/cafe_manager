```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sdf = pd.read_csv('../data/daily_stock_levels.csv')
sdf['Date'] = pd.to_datetime(sdf['Date'])
sdf.head(2)
```
